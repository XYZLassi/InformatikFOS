# SciPy - Diagramme

## Diagramme einfacher mathematischer Funktionen

```python
import numpy as np # Import die Bibliothek Numpy
import matplotlib.pyplot as plt # Importiert die Bibliothek matplotlib


v = 5 # Definiert die Variable v mit einem Wert von 5 m/s
t = np.linspace(0,10) # Definiert die Variable t mit einer Zeit von 0 bis 10 Sekunden


s = v * t # Berechnet den zurück gelegten Weg und speichert die Werte in der Variablen s

plt.plot(t,s) # Zeichnet ein Diagramm mit der Zeit t als X-Achse und s als Y-Achse
```
