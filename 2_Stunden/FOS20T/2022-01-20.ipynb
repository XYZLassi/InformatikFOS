```python
# Variante Holzhammer

def print_summenformel(n):
    summe= 0 # Setzt die Variable auf 0
    for summand in range(n+1): # Schleife von 0 bis 9
        summe = summe + summand # Summiert den aktuellen Summanden auf
    # print(summe)
```

```python
%timeit print_summenformel(100000)
```

```python
# Aufgabe 3

# 3! = 3 * 2 * 1
# n! = n * (n-1) * (n- 2) * ... * 1 

# Aufgabe Programmieren Sie es einfach mal: EA: 10 min

def print_fakultät(n):
    mul= 1 # Setzt die Variable auf 1
    for faktor in range(n):
        mul = (faktor + 1) * mul # Summiert den aktuellen Summanden auf
    return mul # Gibt den Wert zurück 
%timeit print_fakultät(100000)

# Länge Laufzeit durch eine interne Addition
```
