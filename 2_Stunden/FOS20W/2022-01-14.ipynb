```python
# Erstellen Sie eine Funktion:
# Welche ihnen den Schnittpunkt mit der X 
# einer linearen Funktion berechnet
# Funktion: f(x) = y = m * x + n
# EA: 10 min

def schnittpunkt(m,n):
    x_0 =  - n / m # Schnittpunkt
    print(f'Schnittpunkt für 0 = {m}*x + {n}: ({x_0},0)')
    
schnittpunkt(5,3) # f(x) = 5*x +3
schnittpunkt(5,-3) # f(x) = 5*x - 3
schnittpunkt(5,0) # f(x) = 5*
```
